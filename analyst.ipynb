```python
import pandas as pd
import numpy as np
import random
```

```python
data = pd.read_csv('car_data.csv')

#chèn thêm brand và số lượng lượt tìm kiếm

row, col = data.shape
Name = list(data['Name'])
brand = []
Number = []

for i in range(row):
    brand.append(Name[i].split()[0])
    Number.append(random.randint(1,7))
brand = list(brand)
data['Brand'] = brand
data['Number'] = Number

#tạo df cho danh mục brand, giá tiền trung bình người timd kiếm quan tâm, lượt tìm kiếm cho từng brand
 
analistData = pd.DataFrame(columns=['Brand', 'Averange Price', 'Value'])
analistData['Brand'] = list(set(brand))
averange = []
value = []
sumNum = 0
AvgPrice = 0

for i in range(len(set(brand))):
    d = data[data['Brand'] == str(brand[i])]
    d.to_csv()
    listNum = list(d['Number'])
    listPrice = list(d['Price'])
    sumNum = 0
    AvgPrice = 0
    for j in range(len(listNum)):
        sumNum = sumNum + int(listNum[j])
        AvgPrice = listNum[j] * listPrice[j] + AvgPrice
    averange.append(AvgPrice/sumNum)
    value.append(sumNum)
 
analistData['Averange Price'] = averange
analistData['Value'] = value

#tạo  biểu đồ phân tích theo số lượng truy cập

analistData['Value'].plot.bar(xlabel = 'Brand', ylabel = 'Value')

analistData




```
